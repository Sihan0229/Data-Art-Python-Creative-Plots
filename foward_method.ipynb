```python
import numpy as np
import matplotlib.pyplot as plt

#  dy/dx = y
def f(x, y):
    return y

# 
def euler_method(f, x0, y0, h, steps):
    x_values = [x0]
    y_values = [y0]
    for _ in range(steps):
        y0 = y0 + h * f(x0, y0)
        x0 += h
        x_values.append(x0)
        y_values.append(y0)
    return x_values, y_values

def midpoint_method(f, x0, y0, h, steps):
    x_values = [x0]
    y_values = [y0]
    for _ in range(steps):
        k1 = f(x0, y0)
        k2 = f(x0 + h/2, y0 + h/2 * k1)
        y0 = y0 + h * k2
        x0 += h
        x_values.append(x0)
        y_values.append(y0)
    return x_values, y_values

def rk4_method(f, x0, y0, h, steps):
    x_values = [x0]
    y_values = [y0]
    for _ in range(steps):
        k1 = f(x0, y0)
        k2 = f(x0 + h/2, y0 + h/2 * k1)
        k3 = f(x0 + h/2, y0 + h/2 * k2)
        k4 = f(x0 + h, y0 + h * k3)
        y0 = y0 + (h/6) * (k1 + 2*k2 + 2*k3 + k4)
        x0 += h
        x_values.append(x0)
        y_values.append(y0)
    return x_values, y_values

# 
x0, y0 = 0, 1   # 
h = 0.1         # 
steps = 20      # 

# 
x_euler, y_euler = euler_method(f, x0, y0, h, steps)
x_midpoint, y_midpoint = midpoint_method(f, x0, y0, h, steps)
x_rk4, y_rk4 = rk4_method(f, x0, y0, h, steps)

# 
x_true = np.linspace(0, 2, 100)
y_true = np.exp(x_true)

# 
plt.figure(figsize=(10, 6))
plt.plot(x_true, y_true, 'k-', label='True Solution $e^x$')
plt.plot(x_euler, y_euler, 'ro-', label='Euler Method')
plt.plot(x_midpoint, y_midpoint, 'go-', label='Midpoint Method')
plt.plot(x_rk4, y_rk4, 'bo-', label='RK4 Method')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparison of Numerical Methods')
plt.legend()
plt.grid()
plt.show()
```
