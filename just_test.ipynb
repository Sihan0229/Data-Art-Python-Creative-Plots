```python
import matplotlib.pyplot as plt

years = [1, 2, 3, 4, 5]
metrics = {
    '应收账款': [16.67, 11.11, 14.29, 16.00, 26.67],
    '存货': [2.67, 1.61, 2.09, 2.67, 8.89],
    '总资产': [0.30, 0.22, 0.21, 0.23, 0.57]
}

fig, ax1 = plt.subplots(figsize=(10, 6))

# 左侧坐标轴
ax1.plot(*arg: years, metrics['应收账款'], 'b-o', label='应收账款周转率')
ax1.plot(*arg: years, metrics['存货'], 'orange-o', label='存货周转率')
ax1.set_xlabel('年份')
ax1.set_ylabel('周转次数')
ax1.set_ylim(0, 30)

# 右侧坐标轴
ax2 = ax1.twinx()
ax2.plot(years, metrics['总资产'], 'g--s', label='总资产周转率')
ax2.set_ylabel('总资产周转率', color='g')
ax2.set_ylim(0, 0.6)

plt.title('运营能力指标趋势分析(5年期)')
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))
plt.grid(True)
plt.show()
```

```python
from matplotlib import rcParams
import matplotlib.pyplot as plt

years = [1, 2, 3, 4, 5]
metrics = {
    '应收账款': [16.67, 11.11, 14.29, 16.00, 26.67],
    '存货': [2.67, 1.61, 2.09, 2.67, 8.89],
    '总资产': [0.30, 0.22, 0.21, 0.23, 0.57]
}

# 设置中文字体
rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

fig, ax1 = plt.subplots(figsize=(10, 6))

# 左侧坐标轴
ax1.plot(years, metrics['应收账款'], 'b-o', label='应收账款周转率')
ax1.plot(years, metrics['存货'], 'r-o', label='存货周转率')  # 使用十六进制颜色代码
ax1.set_xlabel('年份')
ax1.set_ylabel('周转次数')
ax1.set_ylim(0, 30)

# 右侧坐标轴
ax2 = ax1.twinx()
ax2.plot(years, metrics['总资产'], 'g--s', label='总资产周转率')
ax2.set_ylabel('总资产周转率', color='g')
ax2.set_ylim(0, 0.6)

plt.title('运营能力指标趋势分析(5年期)')
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))
plt.grid(True)
plt.show()
```
